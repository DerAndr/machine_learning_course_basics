## 2.1 Analyzing Categorical Variables

**Example: Species Distribution**

Run this cell to see how to analyze a categorical variable:

## 2.2 Analyzing Numerical Variables

**Example: Body Mass Analysis**

Run this cell to see a complete numerical analysis:

````python
# Remove missing values for cleaner analysis
body_mass = df['body_mass_g'].dropna()

# Calculate key statistics
mean_mass = body_mass.mean()
median_mass = body_mass.median()
std_mass = body_mass.std()
min_mass = body_mass.min()
max_mass = body_mass.max()

print("📊 Body Mass Statistics:")
print(f"  Mean:   {mean_mass:.2f} g")
print(f"  Median: {median_mass:.2f} g")
print(f"  Std Dev: {std_mass:.2f} g")
print(f"  Range:  {min_mass:.0f} - {max_mass:.0f} g")
print(f"  Skewness: {body_mass.skew():.3f}")
````

````python
# Create histogram with mean and median lines
plt.figure(figsize=(10, 6))
plt.hist(body_mass, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_mass, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_mass:.0f}g')
plt.axvline(median_mass, color='green', linestyle='--', linewidth=2, label=f'Median: {median_mass:.0f}g')
plt.title('Distribution of Penguin Body Mass', fontsize=16, fontweight='bold')
plt.xlabel('Body Mass (g)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

# Part 4: Multivariate Analysis - Relationships (20 min)

Let's explore how different measurements relate to each other.

## 4.1 Scatter Plot: Two Variables

**Example: Flipper Length vs Body Mass**

Run this to see the relationship:

````python
# Create scatter plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', 
                hue='species', style='sex', s=100, alpha=0.7,
                palette=['#FF6B35', '#004E89', '#F77F00'])
plt.title('Flipper Length vs Body Mass', fontsize=16, fontweight='bold')
plt.xlabel('Flipper Length (mm)', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Calculate correlation
correlation = df[['flipper_length_mm', 'body_mass_g']].corr().iloc[0, 1]
print(f"\n📊 Correlation: {correlation:.3f}")
print("💡 Strong positive correlation! Longer flippers = heavier penguins")
````

## 4.3 Pair Plot - See All Relationships at Once
This visualization combines every numerical pair so you can spot clusters and species separation in one grid.

````python
# Create a pair plot
pairplot = sns.pairplot(df, vars=numerical_cols, hue='species',
                         palette=['#FF6B35', '#004E89', '#F77F00'],
                         diag_kind='kde', plot_kws={'alpha': 0.6})
pairplot.fig.suptitle('Pair Plot: All Penguin Measurements', y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("💡 This is a powerful visualization!")
print("  • Diagonal = distribution of each variable")
print("  • Off-diagonal = relationships between pairs")
print("  • Colors = different species")
````
