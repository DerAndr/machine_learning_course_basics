## Part 1: The "Guts" of a Neural Network (NumPy)

Before using a framework like PyTorch, let's understand the math of a single neuron.

$$ z = w \cdot x + b $$
$$ y = \sigma(z) $$

### Why not just use NumPy forever?
You might ask: *If we can build this in NumPy, why learn PyTorch/TensorFlow?*

1.  **Autograd (Automatic Differentiation)**: In NumPy, you have to manually derive and write the code for backpropagation (calculating gradients). For huge networks, this is incredibly complex and error-prone. PyTorch does this for you with `loss.backward()`.
2.  **GPU Acceleration**: NumPy runs on CPU. PyTorch Tensors can run on GPU (10x-100x faster), which is essential for Deep Learning.
3.  **Pre-built Layers**: You don't want to type `w*x+b` for a 100-layer network. PyTorch gives you `nn.Linear`, `nn.Conv2d`, etc.

### Task 1.6: The "Nightmare" - A 3-Layer Network
Finally, to prove why we need PyTorch, let's try to conceptualize a network with **one hidden layer** (Input -> Hidden -> Output).

Structure:
*   Input: $x$ (Dimension $1 	imes d_{in}$)
*   Layer 1 (Hidden): $z_1 = x \cdot W_1 + b_1$, $a_1 = \sigma(z_1)$
